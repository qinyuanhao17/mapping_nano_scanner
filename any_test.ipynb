```python
x_list = np.arange(10, 20+0.5, 0.5)
y_list = np.arange(10, 20+0.5, 0.5)
y_list_check_index = list(y_list)
for i in y_list:
    # self.y_task.write(i)
    if y_list_check_index.index(i) % 2 == 0:
        for j in x_list:
            # self.x_task.write(j)
            print('[{},{}]'.format(j,i))
    elif y_list_check_index.index(i) % 2 != 0:
        for j in x_list[::-1]:
            # self.x_task.write(j)
            print('[{},{}]'.format(j,i))
```
